```python
import pandas as pd
import numpy as np
from uszipcode import SearchEngine, SimpleZipcode, Zipcode

file = "C:\\Users\\luis3\\OneDrive\\Escritorio\\COLLEGE\\NJIT\\FALL 2020\\CS301 - Intro to Data Science\\Project\\Motor_Vehicle_Collisions_-_Crashes.csv"
f1 = pd.read_csv(file)
f1.head(100)
```

```python
borough = f1[f1['BOROUGH'].notnull()].reset_index(drop=True)
borough.head(50000)
```

```python
s2 = f1.loc[f1['LATITUDE'].notnull() & f1['LONGITUDE'].notnull() & f1['BOROUGH'].isnull()].reset_index(drop=True)
s2.head(100)
```

```python
counter1 = 0
counter2 = 0
search = SearchEngine(simple_zipcode=True)
for i in range(s2["LOCATION"].shape[0]):
    try:
        zipcode = search.by_coordinates(s2["LATITUDE"][i], s2["LONGITUDE"][i])
        s2["BOROUGH"][i] = zipcode[0].major_city
        s2["ZIP CODE"][i] = zipcode[0].zipcode
        counter1 = counter1 + 1
        #print(counter1+1,'GOOD:::::', s2["LATITUDE"][i],' - ', s2["LONGITUDE"][i])
        
    except:
        counter2 = counter2 + 1
        print(counter2+1, 'BAD::::::::',s2["LATITUDE"][i],' - ', s2["LONGITUDE"][i])
        s2.drop([i])
    
```

```python
s3 = s2.loc[s2['BOROUGH'].notnull()]
s3.shape
```

```python
s4 = pd.concat([borough, s3])
s4.head()
```
